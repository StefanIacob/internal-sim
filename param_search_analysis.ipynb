```python
%matplotlib inline 
#widget
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.5)
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

from matplotlib import cm
```

```python
folder = '29-10-2020'
av_dist_param = np.zeros((5, 6))
for i in range(3,8):
    for j in range(-1, 5):
        runs = []
        for k in range(1,4):
            path = os.path.join(folder, "tau_0.0" + str(i), "delta_"+str(np.round(j*10e-5, decimals=4)), str(k), "fm_error_data.npy")
            data = np.load(path)[:, :2]
            dist = np.sqrt(data[:,0]**2 + data[:,1]**2)
            avg = np.average(dist)
            runs.append(avg)
        av_dist_param[i-3, j+1] = sum(runs)/len(runs)
print(av_dist_param)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(0.03, 0.07, 0.01)

Y = np.arange(-0.0001, 0.0005, 0.0001)
X, Y = np.meshgrid(X, Y)

Z = av_dist_param
# Plot the surface.
surf = ax.plot_surface(X, Y, Z.T, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
fig.show()
```

```python
# plot best runs
folder1 = os.path.join(folder, "tau_0.062", "delta_0.0005")
folder2 = os.path.join(folder, "tau_0.078", "delta_0.0005")

example = np.load(os.path.join(folder1, "1", "fm_error_data.npy"))

def moving_average(a, axis, n=3):
    ret = np.cumsum(a, axis, dtype=float)
    size = a.shape[axis]
    indices1 = np.array(range(n, size))
    indices2 = np.array(range(size-n))
    ret = np.take(ret, indices1, axis=axis) - np.take(ret, indices2, axis=axis)
    size2 = ret.shape[axis]
    indices3 = np.array(range(n-1, size2))
    ret = np.take(ret, indices3, axis=axis)/n
    return ret

smooth = moving_average(example, 0, 10)
plt.figure(figsize=(10,5))
plt.plot(smooth[:,0])
```
